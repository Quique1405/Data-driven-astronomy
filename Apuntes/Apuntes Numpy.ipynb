### Sobre los objetos iterador e iterable:

Definición formal de iterable

En Python, un objeto *iterable* es cualquier objeto que puede devolver un *iterador* cuando se le pasa a la función iter().

```python
A=[1,2,3] #Objeto iterable
B=iter(A) #Iterador
B_=A.__iter__() #Método alternativo
print(B)
print(B_)
```

```python
print(next(B))
print(next(B))
print(next(B))
```

```python
print(next(B)) #Error porque ya acabó la iteración
```

iter(iterable) → crea un nuevo iterador.

iter(iterator) → devuelve el mismo iterador (no reinicia la secuencia).

```python
A=[1,2,3]# A y B son objetos iterables
B=[4,5,6]# 
C=zip(A,B) #C es un iterador (una tupla)
next(C)
```

```python
G=[[1,2,3],[2,3,4]]
b=iter(G)
next(b)
```
