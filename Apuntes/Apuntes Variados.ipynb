## Sobre el tamaño en memoria

sys.getsizeof hace referencia al tamaño de un objeto, bien sea un entero, flotante, lista, etc.

```python
A = 2
B = 67.567
C = True
D = []
E = [1]
for g in  A, B, C, D, E:
    print(sys.getsizeof(g))
```

numpy.nbytes retornará el espacio en memoria de la data contenida en un array, nótese que no es el espacio del array en sí

```python
A= np.array([])
print(type(A), A.nbytes)
print(sys.getsizeof(A))
```
