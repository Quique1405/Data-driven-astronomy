# Apuntes de Git y Github en la terminal

### cls

Limpia la terminal

### git init

Inicia el control de versiones en la carpeta de trabajo, solo se hace una vez

### git version

Devuelve la versión de git que se está usando

### git config username --global user.name "texto"

Configura el username global

### git add . 

Sube todos los cambios al área de preparación

### git commit -m "texto"

Actualiza los cambios en el repositorio local

### git log --oneline

Visualizar todos los commits
